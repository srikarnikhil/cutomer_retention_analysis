### Task 1.2 - Convert the given json data to a csv file

```python
retention_dataframe = pd.read_csv('retention_data.csv')
```

```python
retention_dataframe.head()
```

```python
retention_dataframe_cleaned = retention_dataframe.copy(deep=True)
```

```python
# cleaning age and gross_income columns meanwhile converting them to numeric type Task 1.4a and 1.4b
```

```python
retention_dataframe_cleaned['age'] = retention_dataframe_cleaned['age'].apply(lambda x: int(x.split(" ")[0]))
```

```python
retention_dataframe_cleaned.head()
```

```python
retention_dataframe_cleaned['gross_income'] = retention_dataframe_cleaned['gross_income'].apply(lambda x: float(x.split(" ")[1]))
```

```python
retention_dataframe_cleaned.head()
```
